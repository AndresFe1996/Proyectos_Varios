```python
import requests  ## importar la libreria
```

```python
## para hacer una solicitud get, debemos tener una pagina

url="http://hdeleon.net"  ## colocamos en una variable la direccion web de la pagina
r=requests.get(url)  
data=r.text

print(data) ## como vemos trajo la informacion correctamente

## para hacer una solicitud mas interesante, debemos pedirla sobre un servicio y en este caso vamos a usar
## JSONPLACEHOLDER
```

```python
## esteservicio es un servicio de paginas con informacion falsa, pero es legal porque es usada para probar este tipo
## de funciones request


URL_1="https://jsonplaceholder.typicode.com/posts"
r1=requests.get(URL_1)
data1=r1.text

print(data1)
```

```python
URL_1="https://jsonplaceholder.typicode.com/posts"
r1=requests.get(URL_1)
data1=r1.json()  ### podemos cambair la forma de ver la informacion y convertirla en formato json(libreria) colcoando json

print(data1) ## el resultado es un diccionario que como ya sabemos, podemos convertir en un data frame

## para buscar informacion podemos usar el metodo for

for d in data1:
    print(d["title"])
    print("***********************")
```
