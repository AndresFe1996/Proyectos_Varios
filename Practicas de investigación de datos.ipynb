```python
import pandas as pd
```

```python
# importar la base de datos a analizar
tasas_cap=pd.read_csv("C:/Users/Home/Downloads/Tasas_de_interes_de_captacion_y_operaciones_del_mercado_monetario.csv",
                     parse_dates=["FECHACORTE"])
print(tasas_cap.head())
```

```python
print(type(tasas_cap["FECHACORTE"]))
tasas_cap.info()
```

```python
# filtrar las fechas que vamos a analizar

tasas_cap_filtrado=tasas_cap[(tasas_cap['FECHACORTE'] >= '2021-01-01') & (tasas_cap['FECHACORTE'] <= '2022-12-31')]
tasas_cap_filtrado.tail()
```

```python
# FILTRAR LAS COLUMNAS A REVISAR

tasas_cap_c_filter=tasas_cap_filtrado[["NOMBREENTIDAD","FECHACORTE","DESCRIPCION","TASA","MONTO"]]


"""# agregar comas para separar miles 
tasas_cap_c_filter["MONTO"]=tasas_cap_c_filter["MONTO"].apply('{:n}'.format)
tasas_cap_c_filter"""
```

```python
# CONSOLIDAR LOS MONTOS POR LA DESCRIPCION GROUP BY

tasas_cap_c_filter.groupby('DESCRIPCION').size() # cuantos registros hay por descripcion
```

```python
tasas_cap_c_filter.groupby(["NOMBREENTIDAD",'MONTO','DESCRIPCION']).agg("count") # agregacion: permite colcoar
#mas operaciones estadisticas en una tabla
```

```python
# tambien podemos hacer agg por lista, aqui nos trae la descripcion en sus cuentas (aplica para columna numerica no como esto :( )
tasas_cap_c_filter.groupby(["NOMBREENTIDAD",'MONTO','DESCRIPCION']).agg({"DESCRIPCION":["min","max","count"]})
```

```python
tasas_cap_c_filter.groupby("NOMBREENTIDAD")["TASA"].mean() # AGRUPAR POR UNA COLUMNA ESPECIFICA
```
