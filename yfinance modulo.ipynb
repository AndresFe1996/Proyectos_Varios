```python
import yfinance as yf #importar el modulo yfinance
```

```python
sura=yf.Ticker("PFGRUPSURA.CL")

"""El módulo Ticker, que le permite acceder a los datos de ticker de una manera más pitónica:"""
```

```python
sura.info
```

```python
hist = sura.history(period="1mo") """devuelve un objeto pandas con el historico de la accion"""
print(hist)
```

```python
# show actions (dividends, splits, capital gains)
print(sura.actions)
print(sura.dividends)
sura.splits
sura.capital_gains  # only for mutual funds & etfs
```
